```python
import numpy as np
import matplotlib.pyplot as plt

probabilidad_exito = 0.7
pedidos_por_dia = 20
dias = 7

pedidos_satisfechos = np.random.binomial(pedidos_por_dia, probabilidad_exito, dias)
print("Pedidos satisfechos por día (Distribución Binomial):", pedidos_satisfechos)

plt.figure()
plt.bar(range(1, dias+1), pedidos_satisfechos)
plt.xlabel('Día')
plt.ylabel('Pedidos Satisfechos')
plt.title('Simulación de Pedidos Satisfechos por Día')
plt.show()
```
