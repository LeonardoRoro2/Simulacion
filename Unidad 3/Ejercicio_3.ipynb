```python
import numpy as np
import matplotlib.pyplot as plt

promedio_espera = 10
std_espera = 2
clientes = 50

# Parámetros para lognormal
mu = np.log(promedio_espera / np.sqrt(1 + (std_espera**2 / promedio_espera**2)))
sigma = np.sqrt(np.log(1 + (std_espera**2 / promedio_espera**2)))

# Generar tiempos de espera
tiempos_espera = np.random.lognormal(mu, sigma, clientes)
print("Tiempos de espera generados (Distribución LogNormal):", tiempos_espera)

plt.figure()
plt.hist(tiempos_espera, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Tiempo de Espera (min)')
plt.ylabel('Frecuencia')
plt.title('Simulación de Tiempos de Espera en Servicio Técnico')
plt.show()
```
