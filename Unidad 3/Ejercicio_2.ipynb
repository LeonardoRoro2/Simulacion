```python
import numpy as np
import matplotlib.pyplot as plt

tasa_llegada = 1 / 15  # 1 llegada cada 15 minutos
horas_operacion = 12
minutos_operacion = horas_operacion * 60

# Generar tiempos entre llegadas
tiempos_entre_llegadas = np.random.exponential(scale=1/tasa_llegada, size=minutos_operacion)
tiempo_acumulado = np.cumsum(tiempos_entre_llegadas)
tiempo_acumulado = tiempo_acumulado[tiempo_acumulado <= minutos_operacion]

print("Número de vehículos llegados en 12 horas:", len(tiempo_acumulado))

plt.figure()
plt.hist(tiempos_entre_llegadas, bins=30, density=True)
plt.xlabel('Tiempo entre llegadas (min)')
plt.ylabel('Densidad')
plt.title('Distribución Exponencial - Tiempos entre Llegadas')
plt.show()
```
